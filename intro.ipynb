```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns 
sns.set()
```

```python
#loading data 
data = pd.read_csv("1.01. Simple linear regression.csv")
data.head(20)
```

```python
data.describe()
```

```python
y = data['GPA']
x1 = data['SAT']
```

```python
plt.scatter(x1,y)
plt.xlabel('SAT',fontsize=20)
plt.show()
```

```python
#regression prep
x = sm.add_constant(x1) #this line takes a dataframe (x1) and adds initial column const with 1,0
results = sm.OLS(y,x).fit() #OLS: Ordinary Least Squares - an estimation technique allowing to obtain the fit of the model
results.summary()
```

```python
plt.scatter(x1,y)
yhat = 0.0017*x1 + 0.275
fig = plt.plot(x1,yhat,label='regression line',lw=4,c='red')
plt.show()
```
