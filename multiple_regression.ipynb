```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns 
sns.set()
```

```python
#importing data 
data = pd.read_csv("1.02. Multiple linear regression.csv")
data
```

```python
data['Randomizer'] = data['Rand 1,2,3']
```

```python
del data['Rand 1,2,3']
data
```

```python
data.describe()
```

```python
#creating first multiple regression

y = data['GPA']
x1 = data[['SAT','Randomizer']] # we can perceive x1 as a dataframe containing both x series 
```

```python
x = sm.add_constant(x1)
results = sm.OLS(y,x).fit()
```

```python
results.summary()
```
